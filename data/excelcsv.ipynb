```python
import pandas as pd
def people(cell):
    if cell=="n.a.":
        return "WMT founder"
    else:
        return cell
    
def eps_cell(cell):
    if cell=="not available":
        return None
    return cell 

df=pd.read_excel("stock_data.xlsx",converters={
    'people':people,
    'eps':eps_cell
})
df
```

```python
import pandas as pd
df_stocks=pd.DataFrame({
    'tickers':["Google","TATA","RIL"],
    'price':[45,1023,65],
    'people':["larry page","ratan tata","mukhesh ambani"],
})


df_weather=pd.DataFrame({
    'date':['1/1/2020','2/1/2020','3/1/2020'],
    'temparature':[12,13,14],
    'event':['rainy','foggy','hot']
})


####excel write combining all the two dataframes
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer,sheet_name="stocks",index=False)
    df_weather.to_excel(writer,sheet_name="weather")
```
